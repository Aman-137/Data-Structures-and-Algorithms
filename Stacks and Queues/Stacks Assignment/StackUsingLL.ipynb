```python
# Stack using Linked list.



from sys import stdin

class Node :

	def __init__(self, data) :
		self.data = data
		self.next = None
        


'''----------------- Public Functions of Stack -----------------'''


        
        
        
        
class Stack :

	def __init__(self) :
		self.__head = None
		self.__cnt = 0


	def getSize(self) :
		return self.__cnt


	def isEmpty(self) :
		return self.__cnt == 0


	def push(self, data) :
		newNode = Node(data)

		if self.__head is None :
			self.__head = newNode

		else :
			newNode.next = self.__head
			self.__head= newNode

		self.__cnt += 1


	def pop(self) :
		if self.__head is None :
			return -1

		ans = self.__head.data

		self.__head = self.__head.next
		self.__cnt -= 1

		return ans


	def top(self) :
		if self.__head is None :
			return -1

		return self.__head.data
		


        
        
        
        
#main
q = int(stdin.readline().strip())

stack = Stack()

while q > 0 :

	inputs = stdin.readline().strip().split(" ")
	choice = int(inputs[0])

	if choice == 1 :
		data = int(inputs[1])
		stack.push(data)

	elif choice == 2 :
		print(stack.pop())

	elif choice == 3 :
		print(stack.top())

	elif choice == 4 : 
		print(stack.getSize())

	else :
		if stack.isEmpty() :
			print("true")
		else :
			print("false")

	q -= 1
```

```python
# Balanced Paranthesis 



from sys import stdin


def isBalanced(expression) :
	#Your code goes her
    s = []
    
    for char in expression:
        
        if char in '({[':
            s.append(char)
            
        elif char == ')':
            
            if (not s or s[-1] != '('):
                return False
            s.pop()
            
        elif char == '}':
            
            if (not s or s[-1] != '{'):
                return False
            s.pop()
            
        elif char == ']':
            
            if (not s or s[-1] != '['):
                return False
            s.pop()
            
    if (not s):
        return True
    else:
        return False



	




#main
expression = stdin.readline().strip()

if isBalanced(expression) :
	print("true")

else :
	print("false")
```

```python
# Reverse Stack.



from sys import stdin, setrecursionlimit
setrecursionlimit(10 ** 6)


def reverseStack(inputStack, extraStack) :
	#Your code goes here
    if (len(inputStack) <= 1):
        return
    
    while (len(inputStack) != 1):
        
        elem = inputStack.pop()
        extraStack.append(elem)
        
    lastElement = inputStack.pop()
    
    while (len(extraStack) != 0):
                
        elem = extraStack.pop()
        inputStack.append(elem)
        
    
    reverseStack(inputStack, extraStack)
    
    inputStack.append(lastElement)
        
    






'''-------------- Utility Functions --------------'''
def isEmpty(s): # s is a list
    return len(s)==0

#Taking input using fast I/o method
def takeInput() :
	size = int(stdin.readline().strip())
	inputStack = list()

	if size == 0 :
		return inputStack


	values = list(map(int, stdin.readline().strip().split(" ")))
	inputStack = values

	return inputStack


#main
inputStack = takeInput()
extraStack = list()

reverseStack(inputStack, extraStack)

while not isEmpty(inputStack) :
	print(inputStack.pop(), end = " ")
```

```python
# Minimum Brackets Reversal.



from sys import stdin

def countBracketReversals(inputString) :
    # Your code goes here
    if len(inputString) % 2:
        return -1
    
    inversions = 0
    open = 0
    
    for i in range(len(inputString)):
        
        if inputString[i] == '{':
            
            open += 1
        
        else:
            if open:
                open = open - 1
                
            else:
                inversions += 1
                open = 1
                
                
                
    return inversions + open // 2
                






#main
print(countBracketReversals(stdin.readline().strip()))
```
